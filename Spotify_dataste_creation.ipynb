```python
from sklearn.cluster import KMeans
import pandas as pd
```

```python
df = pd.read_csv('songdataset_withmood.csv')
df.shape
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
X = df.drop(['track_id', 'artist_name', 'track_name','popularity','track_type', 'uri', 'track_href', 'analysis_url', 'duration_ms',
       'time_signature','instrumentalness','genre','key','mode'], axis = 1)
```

```python
X
```

```python
kc = KMeans(n_clusters=4, init='k-means++')
classes = list(kc.fit_predict(X))
for i in range(len(classes)):
    if classes[i] == 0:
        classes[i]='sad' 
    elif classes[i] == 1:
        classes[i]='angry'
    elif classes[i] == 2:
        classes[i]='happy'
    elif classes[i] == 3:
        classes[i]='surprised'
```

```python
mood_song = df
mood_song['mood'] = classes
mood_song
```
